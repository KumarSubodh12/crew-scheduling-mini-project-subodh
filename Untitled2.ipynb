## Backtracking - Crew Scheduling

### Explanation
Backtracking is used to assign flights to crew members while satisfying constraints like no overlapping schedules.

```python
flights = [('F1', 9, 11), ('F2', 10, 12), ('F3', 12, 14)]
crew = ['C1', 'C2']

def is_valid(schedule, crew_member, flight):
    for f in schedule.get(crew_member, []):
        if not (flight[2] <= f[1] or flight[1] >= f[2]):
            return False
    return True

def assign(flights, crew, schedule={}):
    if not flights:
        return schedule

    flight = flights[0]

    for c in crew:
        if is_valid(schedule, c, flight):
            schedule.setdefault(c, []).append(flight)
            result = assign(flights[1:], crew, schedule)
            if result:
                return result
            schedule[c].pop()

    return None

print("Schedule:", assign(flights, crew))
```

## Branch and Bound

### Explanation
Branch and Bound improves backtracking by pruning paths that exceed cost limits.

```python
def branch_and_bound(flights, crew):
    best = None

    def helper(i, schedule):
        nonlocal best

        if i == len(flights):
            best = schedule.copy()
            return

        for c in crew:
            if is_valid(schedule, c, flights[i]):
                schedule.setdefault(c, []).append(flights[i])
                helper(i+1, schedule)
                schedule[c].pop()

    helper(0, {})
    return best

print("Optimized Schedule:", branch_and_bound(flights, crew))
```

## String Matching - Naive Algorithm

```python
def naive_search(text, pattern):
    n = len(text)
    m = len(pattern)
    count = 0

    for i in range(n - m + 1):
        match = True
        for j in range(m):
            count += 1
            if text[i+j] != pattern[j]:
                match = False
                break
        if match:
            print("Pattern found at index", i)

    print("Comparisons:", count)

text = "ABABDABACDABABCABAB"
pattern = "ABABCABAB"

naive_search(text, pattern)
```

```python
import time

text = "A" * 1000 + "B"
pattern = "AB"

start = time.time()
naive_search(text, pattern)
end = time.time()

print("Execution Time:", end - start)
```
